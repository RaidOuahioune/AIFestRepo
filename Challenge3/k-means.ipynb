#### Targerting Treatment For Heart Disease Patients

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

heart_disease = pd.read_csv("heart_disease.csv")
ids=heart_disease.id
heart_disease = heart_disease.drop("id", axis=1)


# Print the first ten rows of the data set
print(heart_disease.head(10))

# Check that only numeric variables
print(heart_disease.dtypes)
```

#### Data Cleaning

```python
# Replace "male" with 1 and "female" with 0 in the "sex" column
heart_disease['sex'] = heart_disease['sex'].replace({'Male': 1, 'Female': 0})
heart_disease['exang'] = heart_disease['exang'].replace({False: 0, True: 0})
heart_disease['fbs'] = heart_disease['fbs'].replace({False: 0, True: 0})
heart_disease
```

```python
# Drop a specific column
heart_disease = heart_disease.drop("dataset", axis=1)
heart_disease = heart_disease.drop("thal", axis=1)
numeric_col=heart_disease.select_dtypes(include=np.number)
```

```python

# Assuming heart_disease is your DataFrame
column_to_convert = 'cp'

# Convert the values in the specified column to integers
labels, uniques = pd.factorize(heart_disease[column_to_convert])
heart_disease[column_to_convert] = labels + 1

# Print the updated dataset
print(heart_disease[column_to_convert].unique())
```

```python

# Assuming heart_disease is your DataFrame
column_to_convert = 'slope'

# Convert the values in the specified column to integers
labels, uniques = pd.factorize(heart_disease[column_to_convert])
heart_disease[column_to_convert] = labels + 1

# Print the updated dataset
print(heart_disease[column_to_convert].unique())
```

```python

# Assuming heart_disease is your DataFrame
column_to_convert = 'restecg'

# Convert the values in the specified column to integers
labels, uniques = pd.factorize(heart_disease[column_to_convert])
heart_disease[column_to_convert] = labels + 1

# Print the updated dataset
print(heart_disease[column_to_convert].unique())
```

```python
heart_disease=heart_disease.drop("ca",axis=1)
heart_disease
```

```python

# Scaling data and saving as a data frame
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(heart_disease)
scaled = pd.DataFrame(scaled, columns=heart_disease.columns)

# What does data look like now?
print(scaled.describe())
```

#### Grouping Patients

```python
for col in scaled.columns:
    mean_A = scaled[col].mean()
    scaled[col].fillna(mean_A, inplace=True)



import numpy as np
import pandas as pd

np.random.seed(33)

# Select a number of clusters
k = 1

# Run the k-means algorithm
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=k, random_state=30)
first_clust = kmeans.fit(scaled)


# How many patients are in each group?
labels, counts = np.unique(first_clust.labels_, return_counts=True)
cluster_sizes = pd.Series(counts, index=labels)
print(cluster_sizes)

```

#### Another round of k-means

```python
np.random.seed(38)

# Run the k-means algorithm
from sklearn.cluster import KMeans

k = 3
kmeans = KMeans(n_clusters=k, random_state=38)
second_clust = kmeans.fit(scaled)

# How many patients are in each group?
labels, counts = np.unique(second_clust.labels_, return_counts=True)
cluster_sizes = pd.Series(counts, index=labels)
print(second_clust.labels_)
```

From the figure below, we decided to fix k to 1 since all points seems to be clustered together

#### Comparing patient clusters

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Adding cluster assignments to the data
heart_disease['first_clust'] = first_clust.labels_
heart_disease['second_clust'] = second_clust.labels_

# Creating the plots of age and chol for the first clustering algorithm
plot_one = heart_disease.plot.scatter(x='age', y='chol', c='first_clust', colormap='viridis')
plt.show()

# Creating the plots of age and chol for the second clustering algorithm
plot_two = heart_disease.plot.scatter(x='age', y='chol', c='second_clust', colormap='viridis')
plt.show()
```

#### Hierarchical clustering

```python
# Executing hierarchical clustering with complete linkage
from scipy.cluster.hierarchy import linkage, dendrogram

hier_clust_1 = linkage(scaled, method="complete")

# Printing the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(hier_clust_1)
plt.show()

# Getting cluster assignments based on the number of selected clusters
from scipy.cluster.hierarchy import fcluster

k = 3
hc_1_assign = fcluster(hier_clust_1, k, criterion='maxclust')
print(hc_1_assign)
```

#### Hierarchical clustering Another Round

```python
# Executing hierarchical clustering with single linkage
hier_clust_2 = linkage(scaled, method="single")

# Printing the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(hier_clust_2)
plt.show()

# Getting cluster assignments based on the number of selected clusters
hc_2_assign = fcluster(hier_clust_2, k, criterion='maxclust')
print(hc_2_assign)
```

#### Comparing clustering results

```python
# Adding assignments of chosen hierarchical linkage
heart_disease['hc_clust'] = hc_1_assign

# Remove 'sex', 'first_clust', and 'second_clust' variables
hd_simple = heart_disease.drop(['sex', 'first_clust', 'second_clust'], axis=1)

# Getting mean and standard deviation summary statistics
clust_summary = hd_simple.groupby('hc_clust').agg(['mean', 'std'])
clust_summary = clust_summary.reset_index()
print(clust_summary)
```

#### Visualizing the cluster contents

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting age and chol
plot_one = sns.scatterplot(data=heart_disease, x='age', y='chol', hue='hc_clust')
plt.show()

# Plotting oldpeak and trestbps
plot_two = sns.scatterplot(data=heart_disease, x='oldpeak', y='trestbps', hue='hc_clust')
plt.show()
```
